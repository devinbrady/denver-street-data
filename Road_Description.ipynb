# Road Description

Does the crash dataset have the values to tell us whether a crash occurred on a freeway or not? Ideally, I'd like to keep the freeway crashes out of the analysis. 

```python
import pandas as pd
import sqlite3 as sq
```

```python
conn = sq.connect('data/denver_crashes.sqlite')
```

```python
query = """
select 
ROAD_DESCRIPTION
, count(*)

from crashes_raw

group by 1 order by 2 desc
"""

df = pd.read_sql(query, conn)
```
